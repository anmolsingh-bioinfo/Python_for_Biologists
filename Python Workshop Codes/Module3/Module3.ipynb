```python
#


# creating a machine
def add_numbers(number1, number2): # machine declaration, machine name, machine input

    result = number1 + number2 # machine operation / function

    return result # return the output from the machine


# using the created machine - give INPUT & expect OUTPUT
number1 = 150 # which im planning to give in machine
number2 = 190
number3 = 120

output_result1 = add_numbers(number1, number2) # using the machine
output_result2 = add_numbers(number1, number3) # using the machine

print("Output from machine ---",output_result1, output_result2)
print("Input to machine ---", number1, number2)
```

```python
# Necleotide enrichment

dna_seq = "ATGCGTNACG"

base_count = { 'A': 0, 'T': 0, 'G': 0, 'C': 0}

for base in dna_seq:

    if base not in ['A', 'T','G','C']: continue
    base_count[base] += 1

print(base_count)

```

```python
# example with count no. of N's in ATGC's

DNA_Seq = 'GGATGGCCGANCGTTGACNGGAAGNTGAAGTAA'

N_count = 0

for i in DNA_Seq:

    if i == 'N':
        N_count += 1

print('N_count--',N_count)
```

```python
# Reverse compliment

dna_seq = "ATCGTAGCTAAGGCTT"

# complementary strand
def complement(base):
    mapping = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return mapping[base]

complementary = list(map(complement, dna_seq))
print("Complementary strand:", ''.join(complementary))
```

```python
# Keep only purines which are 'A' and 'G'

dna_seq = "ATCGTAGCTAAGGCTT"

def is_purine(base):
    return base in ('A', 'G')

purines = list(filter(is_purine, dna_seq))
print("Purines only:", ''.join(purines), purines)
```
